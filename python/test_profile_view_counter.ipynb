```python
# Testing a profile view counter
import requests


URL = "https://camo.githubusercontent.com/ceb7faeb7a45a2abc72dbbf013b192c762260c1dd33c774d10f91a9a7c8ee139/68747470733a2f2f6b6f6d617265762e636f6d2f67687076632f3f757365726e616d653d73696d656f6e6f6e7365637572697479"


def get_profile_view_count(url: str) -> int:
    """Get profile view count"""
    r = requests.get(url)
    r.raise_for_status()
    text = r.text
    return int(text.split("</text>")[-2].split(">")[-1].replace(",", ""))


# Send 10 requests synchronously and print the result
for _ in range(int(1e3)):
    print(get_profile_view_count(URL))
```
