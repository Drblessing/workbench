```python
s = "aabbc"
t = "abcde"
from collections import Counter

cs = Counter(s)
ct = Counter(t)

# Check if isomorphic
if len(cs) != len(ct):
    print("Not isomorphic")
else:  # Check if the frequency of characters is the same
    if sorted(cs.values()) == sorted(ct.values()):
        print("Isomorphic")
    else:
        print("Not isomorphic")
```
